# Data cleaning

# Average rating per year

```python
rateperyear = data['Rating'].groupby(data['Year']).mean().reset_index()
```

```python
rateperyear
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.plot(rateperyear['Year'],rateperyear['Rating'],marker = 'o',linestyle='-',color = 'b',label = 'Mean Rating')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Mean Rating',fontsize=12)
plt.title('Average Rating per year',fontsize=14)
plt.grid(True,linestyle='-',alpha=0.5)
plt.legend(fontsize=7)
```

# top 10 directors with high grossworldwide 

```python
direct_gross = data['grossWorldWide'].groupby(data['directors']).sum().reset_index()
```

```python
direct_gross
```

```python
direct_gross['directors'] = direct_gross['directors'].astype('string')
```

```python
direct_gross.info()
```

```python
direct_gross['grossWorldWide'].duplicated()
```

```python
#removing duplicates from grossworldwide
direct_gross = direct_gross[direct_gross['grossWorldWide']!=0].reset_index(drop=True)
```

```python
direct_gross
```

```python
#only looking for top10 richest directors since 60yrs
top10 = direct_gross.sort_values('grossWorldWide',ascending=False).head(10).reset_index()
```

```python
top10
```

```python
fig = plt.figure(figsize=(50,10))
```

```python
plt.bar(top10.index,top10['grossWorldWide'],color='skyblue')
plt.xlabel('Directors top10')
plt.ylabel('gross worldwide generated')
plt.title('Rich directors list')
```

# top10 highest grossed movies of all time

```python
data.head()
```

```python
#on basis of rating we can get top10 movies of all time
high_rated = data.sort_values('Rating',ascending=False).head(10).reset_index(allow_duplicates=False)
```

```python
high_rated.drop(columns=['id','index'])
```

```python
high_rated.info()
```

```python
high_rated['Title'] = high_rated['Title'].astype('string')
```

```python
high_rated.info()
```

```python
#selecting columns of dataframe for plotting
high_grossed = high_rated[['Title','Rating','grossWorldWide']]
```

```python
high_grossed
```

```python
high_grossed.fillna(0000.0)
```

```python
#plotting
fig = plt.figure(figsize=(25,10))
```

```python
plt.plot(high_grossed.index,high_grossed['Rating'],marker='o',color='g')
plt.xlabel('indices of Movies',fontsize=15)
plt.ylabel('Ratings',fontsize=15)
plt.title('Top10 highest rated Movies',fontsize=17)
plt.grid(linestyle='-',color='gray')
```

```python
#now,top10 highest grossed movies of all time
fig = plt.figure(figsize=(15,10))
```

```python
plt.plot(high_grossed.index,high_grossed['grossWorldWide'],linestyle='--',color='r',marker='o')
plt.grid(linestyle='-',color='k')
plt.xlabel('indices of movies')
plt.ylabel('grossWorldWide of movies')
plt.title('Top10 highest grossed Movies of all time')
```
